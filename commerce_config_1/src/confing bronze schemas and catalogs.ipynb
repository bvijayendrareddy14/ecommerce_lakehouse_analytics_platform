# Default notebook

This default notebook is executed using a Lakeflow job as defined in resources/sample_job.job.yml.

```python
spark.sql(f"""create schema if not exists {catalog}.{schema} Managed Location 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/' """)
```

```python
# Set default catalog and schema
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.customers (
    customer_id STRING,
    customer_name STRING,
    email STRING,
    city STRING,
    state STRING,
    signup_date DATE,
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/customers';
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.products (
    product_id STRING,
    product_name STRING,
    category STRING,
    brand STRING,
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/products';
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.orders (
    order_id STRING,
    customer_id STRING,
    order_date DATE,
    order_status STRING,
    payment_method STRING,
    order_channel STRING,
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/orders';
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.order_items (
    order_id STRING,
    product_id STRING,
    quantity INT,
    unit_price DECIMAL(10,2),
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/order_items';
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.returns (
    return_id STRING,
    order_id STRING,
    product_id STRING,
    return_date DATE,
    refund_amount DECIMAL(10,2),
    ingestion_time TIMESTAMP,
    source_file STRING
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/returns';
""")
```
