# Default notebook

This default notebook is executed using a Lakeflow job as defined in resources/sample_job.job.yml.

```python
# Set default catalog and schema
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

```python
spark.sql(f"""create schema if not exists {catalog}.{schema }
          Managed Location 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/'
          """)
```

```python
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
spark.sql(f"""
CREATE TABLE if not exists {catalog}.{schema}.customer_stage
(
    customer_id BIGINT,
    customer_name STRING,
    email STRING,
    phone STRING,
    city STRING,
    state STRING,
    country STRING,
    registration_date DATE,
    created_ts TIMESTAMP,
    updated_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/customers'
          """)
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.product_stage
(
    product_id BIGINT,
    product_name STRING,
    category STRING,
    brand STRING,
    unit_price DECIMAL(10,2),
    created_ts TIMESTAMP,
    updated_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/product_stage'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.order_header_stage
(
    order_id BIGINT,
    customer_id BIGINT,
    order_date DATE,
    order_status STRING,
    payment_method STRING,
    total_amount DECIMAL(12,2),
    created_ts TIMESTAMP,
    updated_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/order_header_stage'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.order_detail_stage
(
    order_item_id BIGINT,
    order_id BIGINT,
    product_id BIGINT,
    quantity INT,
    unit_price DECIMAL(10,2),
    line_amount DECIMAL(12,2),
    created_ts TIMESTAMP,
    updated_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/order_detail_stage'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.return_stage
(
    return_id BIGINT,
    order_id BIGINT,
    product_id BIGINT,
    return_date DATE,
    return_reason STRING,
    refund_amount DECIMAL(12,2),
    created_ts TIMESTAMP,
    updated_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/return_stage'
""")
```
