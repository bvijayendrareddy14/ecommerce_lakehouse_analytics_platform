```python
dbutils.widgets.text("catalog", "commerce_main_dev")
dbutils.widgets.text("schema", "gold")
```

```python
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

```python
spark.sql(f"""create schema if not exists {catalog}.{schema}
          MANAGED LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/'
          """)
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.dim_customer
(
    customer_id BIGINT,
    customer_name STRING,
    email STRING,
    city STRING,
    state STRING,
    country STRING,
    registration_date DATE,
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/dim_customer'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.dim_product
(
    product_id BIGINT,
    product_name STRING,
    category STRING,
    brand STRING,
    unit_price DECIMAL(10,2),
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/dim_product'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.fact_sales
(
    order_id BIGINT,
    customer_id BIGINT,
    product_id BIGINT,
    order_date DATE,
    quantity INT,
    unit_price DECIMAL(10,2),
    sales_amount DECIMAL(12,2),
    order_status STRING,
    payment_method STRING,
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/fact_sales'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.fact_returns
(
    return_id BIGINT,
    order_id BIGINT,
    product_id BIGINT,
    return_date DATE,
    refund_amount DECIMAL(12,2),
    return_reason STRING,
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/fact_returns'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.customer_sales_summary
(
    customer_id BIGINT,
    total_orders BIGINT,
    total_quantity BIGINT,
    total_sales DECIMAL(14,2),
    last_order_date DATE,
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/customer_sales_summary'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.product_performance_summary
(
    product_id BIGINT,
    total_orders BIGINT,
    total_quantity_sold BIGINT,
    total_sales DECIMAL(14,2),
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/product_performance_summary'
""")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.{schema}.dim_date
(
    date_key INT,
    full_date DATE,
    day_of_month INT,
    day_name STRING,
    week_of_year INT,
    month_number INT,
    month_name STRING,
    quarter INT,
    year INT,
    is_weekend BOOLEAN,
    created_ts TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/{schema}/dim_date'
""")
```

```python
spark.sql(f"""create schema if not exists {catalog}.util
          MANAGED LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/'
          """)
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {catalog}.util.run_log
(
    run_id STRING,
    pipeline_name STRING,
    task_name STRING,
    status STRING,
    message STRING,
    error_message STRING,
    event_timestamp TIMESTAMP
)
USING DELTA
LOCATION 'abfss://ecommerce@dataprojectadls.dfs.core.windows.net/{catalog}/util/run_log'
""")
```
